```python
from scipy import stats
import numpy as np
import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'
import matplotlib.pyplot as plt
```

```python
arr = np.random.normal(0, 0.1, (1000))
# arr
```

```python
np.median(arr), stats.scoreatpercentile(arr, 50)
```

```python
hist, bins = np.histogram(arr, bins=100)
bins, hist
```

```python
fig, ax = plt.subplots()

ax.hist(arr, bins=100, label='histogram')

# 取区间均值
bins = 0.5 * (bins[:-1] + bins[1:])
b = stats.norm.pdf(bins)
ax1 = ax.twinx() # 沿x轴获取镜像
ax1.plot(bins, b, label='pdf',color='g')

ax.set_ylabel('histogram')
ax1.set_ylabel('probability density function')
# 合并图例
lines, labels = [], []
for axsub in (ax, ax1):
    line, label = axsub.get_legend_handles_labels()
    lines.extend(line)
    labels.extend(label)

plt.legend(lines, labels)  # plt 默认对最后一个子图操作， fig 默认对画布操作

# plt.savefig('plt.png')  # 该方式正确
plt.tight_layout() # 自动布局
plt.show()
fig.savefig('fig.eps')
fig.savefig('fig.svg') # eps 矢量图不支持透明通道，采用svg格式
fig.savefig('fig.png',dpi=1080)
# plt.savefig('plt.png')  # error: 调用 plt.show() 会将 plt 指向新空白图
```

```python
fig, axes = plt.subplots(1, 2, sharex=True) # 在Jupyter中，作用域内会输出一次
axes[0].hist(arr, bins=100, label='histogram')
axes[1].plot(bins, b, label='pdf',color='g')

axes[1].yaxis.tick_right() # 将y标轴放于右方
for ax in axes:
    ax.legend()
# plt.subplots_adjust(wspace=0.3) # 30%比例的宽度间距
plt.tight_layout()  # 调制子图间距
plt.show()
fig.savefig('fig2.png', dpi=1080)
```

```python
fig, axes = plt.subplots(1, 2, sharex=True)
axes[0].hist(arr, bins=100, label='histogram')
axes[1].plot(bins, b, label='pdf',color='g')

axes[1].yaxis.tick_right() # 将y标轴放于右方

# 创建统一图例
lines, labels = [], []
for ax in axes:
    axLine, axLabel = ax.get_legend_handles_labels()
    lines.extend(axLine)
    labels.extend(axLabel)
fig.legend(lines, labels)

plt.tight_layout()  # 调制子图间距
plt.show()
```
