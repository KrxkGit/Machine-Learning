```python
import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'
import matplotlib.pyplot as plt
import numpy as np
from IPython import display
```

```python
# 方式一
fig, ax = plt.subplots()
y1 = []
for i in range(20):
    y1.append(i)  # 每迭代一次，将i放入y1中画出来
    
    ax.cla() # 清理坐标对象内容，为重新绘图做准备
    
    ax.bar(y1, label='test', height=y1, width=0.3)
    
    ax.legend()
    ax.set_xlabel('arr')
    ax.set_ylabel('res')
    ax.set_title('Dynamic_graph')
    ax.grid() # 显示网格
    
    display.display(fig) # 显示图像
#     plt.pause(0.1) # 稍作暂停
    display.clear_output(wait=True)
    
# 保存结果 
fig.savefig('result.jpg')
```
