```python
import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# from mpl_toolkits.mplot3d import Axes3D
```

```python
data = (np.arange(30) + np.random.randn(30) * 10).reshape((-1 ,3))
data2 = (np.arange(30)).reshape((-1 ,3))
# data, data
x, y, z = [data[:, i] for i in range(3)]
x1, y1, z1 = [data2[:, i] for i in range(3)]
```

```python
fig = plt.figure()
ax = fig.add_subplot(121 ,projection='3d')
ax.scatter(x, y, z, c='r', label='random')
ax.scatter(x1, y1, z1, c='g', label='arange')
# ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title(('x - y - z').upper())

ax = fig.add_subplot(122, projection='3d')
ax.scatter(x, y, z, c='r', label='random')
ax.scatter(x1, y1, z1, c='g', label='arange')
# ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title(('x - y - z').upper())

fig.legend(*(ax.get_legend_handles_labels()))
fig.subplots_adjust(wspace=0.5)
fig.savefig('3d.svg')
plt.show()
```
