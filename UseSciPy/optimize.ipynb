```python
from scipy import optimize
import numpy as np
import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'
import matplotlib.pyplot as plt
```

```python
x = np.arange(-10, 10, 0.5)

noise = np.random.randn(1) * 3

def f(x):
    return x**2 + 10 * np.sin(x) - noise

h0 = optimize.fmin_bfgs(f, 5)
h1 = optimize.fmin_bfgs(f, 0)
h2 = optimize.fmin_bfgs(f, -9)
zero = optimize.fsolve(f, 0)

plt.xlabel('x')
plt.ylabel('y')
plt.title('f(x) & f_min')
plt.plot(x, f(x), 'b',label='f(x)')
plt.plot(h0, f(h0), 'ro',label='h0')
plt.plot(h1, f(h1), 'r*',label='h1')
plt.plot(h2, f(h2), 'r^',label='h2')
plt.plot(zero, f(zero), 'go', label='zero')
plt.legend()
```
