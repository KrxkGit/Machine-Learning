```python
import os
import pickle
```

```python
root_path = 'F:/DP-Dataset/dataset-text-formula/'
label_path = root_path + 'dev/labels'
path_split = label_path.split('/')
out_path =  root_path + path_split[-3].replace('dataset-','') + "_" + path_split[-2] + "_out" 
```

```python
out_path
```

```python
file_list = os.listdir(label_path)
```

```python
save_cwd =  os.getcwd()
os.chdir(label_path) # 修改工作目录
labels_name = [] # 用于保存结果
labels_content = []

count = 0
print(f'共{len(file_list)}个')
for file_item in file_list:
    print(f'Reading {count}')
    labels_name.append(file_item)
    with open(file_item, 'r', encoding='utf-8') as f:
        label = f.read()
    
    count += 1
    labels_content.append(label)
    
labels = {'id': labels_name, 'label': labels_content}
os.chdir(save_cwd)
```

```python
labels_name[-1]
```

```python
# 写入
with open(out_path, 'wb') as f:
    pickle.dump(labels, f)
```

```python
# 读出
with open(out_path, 'rb') as f:
    labels_load = pickle.load(f)
labels_load
```
