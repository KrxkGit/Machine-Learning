```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'
```

```python
src_path = 'D:/数学建模/数模第三次作业/筛选结果.xlsx'
dest_path = 'D:/数学建模/数模第三次作业/Compose_selected.xlsx'
```

```python
data = pd.read_excel(src_path, sheet_name='订货', header=0)
data[:10]
```

```python
data2 = pd.read_excel(src_path, sheet_name='供货', header=0)
data2[:10]
```

```python
def preProgress(data, data_compose):
    data_compose.insert(0, '供应商ID', data.iloc[:, 0]) # 复制供应商ID
    data_compose.insert(1, '材料分类', data.iloc[:, 1]) # 复制供应商ID
    group_members = 24
    for i in range(1, 11):
        v = data.iloc[: , 2 + (i-1)*group_members: 2 + i*group_members].sum(axis = 1)
        col_name = "第%d周期" % i
        data_compose.insert(loc=data_compose.shape[1], column=col_name, value=v)
```

```python
data_compose_order = pd.DataFrame()
data_compose_supply = pd.DataFrame()
```

```python
preProgress(data, data_compose_order)
preProgress(data2, data_compose_supply)
```

```python
# 预览
data_compose_order[:10], data_compose_supply[:10]
```

```python
excel_writer = pd.ExcelWriter(dest_path)
data_compose_order.to_excel(excel_writer, '企业的订货量（m³） 合成', index=None)
data_compose_supply.to_excel(excel_writer, '供应商的供货量（m³）合成', index=None)
excel_writer.save()
```
