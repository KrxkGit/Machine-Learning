```python
import pandas as pd
import numpy as np
```

```python
fpath = 'D:/数学建模/数模第四次作业/1数据处理和模型初步建立/total_comment.xlsx'
```

```python
data = pd.read_excel(fpath, header=0)
data.columns
```

```python
data.iloc[:10]
```

```python
data['ID'] = range(1, 1 + data.shape[0])
```

```python
data.iloc[:10]
```

```python
def replaceEv(x):
    ev_map = ["不合理", "较不合理", "一般","较合理", "合理"]
    return ev_map[x - 1]



for column in data.columns:
    if column == 'ID':
        continue
    else:
        data[column] = data[column].map(replaceEv)
```

```python
data.iloc[:10]
```

```python
data.to_excel('D:/数学建模/数模第四次作业/1数据处理和模型初步建立/total_comment_Replace.xlsx')
```
