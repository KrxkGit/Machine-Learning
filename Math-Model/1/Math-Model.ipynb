```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'
```

```python
df_path = 'D:\\数学建模\\数模第四次作业\\A题-附件\\附件2 慢性病及相关因素流调数据.xlsx'
```

```python
data = pd.read_excel(df_path, header=1)
```

```python
data = data.dropna(how='all')
```

```python
data['出生年'] = 2023 - data['出生年']
data = data.rename(columns={'出生年':'年龄'})
```

```python
data.insert(data.shape[1], '吸烟', 0)
data
```

```python
def replace(x):
    if x == 1:
        return "是" 
    elif x == 2: 
        return "已戒烟"
    else:
        return "否"
    
data['吸烟'] = data['是否吸烟'].map(replace)
```

```python
def replace2(x):
    if x == 1:
        return "是"
    elif x == 2: 
        return "否"
    else:
        return "已戒酒"
data.insert(data.shape[1], '饮酒', 0)
data['饮酒'] = data['是否饮酒'].map(replace2)
data
```

```python
data.reset_index()
data['ID'] = range(1, len(data) + 1)
```

```python
data.columns
```

```python
data['职业'].fillna(value=0, inplace=True)
```

```python
data
```

```python
def replaceJob(x):
    jobs = ['其他','工人','农民','军人','行政干部','科技人员','医护人员', '教师',
            '金融财务','商业服务人员','家庭妇女','离、退休人员','待业','学生']
    return jobs[int(x)]

data['职业'] = data['职业'].map(replaceJob)
```

```python
data[data['职业'] == '离、退休人员']['年龄']
```

```python
def replaceAge(x):
    if x in range(0, 7):
        return "婴幼儿"
    elif x in range(7, 13):
        return "少儿"
    elif x in range(13, 18):
        return "青少年"
    elif x in range(18, 46):
        return "青年"
    elif x in range(46, 70):
        return "中年"
    else:
        return "老年"

data.insert(data.shape[1],'年龄阶段',"")
data['年龄阶段'] = data['年龄'].map(replaceAge)
```

```python
data['年龄阶段']
```

```python
def replaceSport(x):
    if x == 1:
        return "轻度"
    elif x == 2:
        return "中度"
    elif x == 3:
        return "重度"

data_JobHomeSport = pd.DataFrame(data, columns=['ID', '工作主要属于以下何种活动', '您做休闲、家务活动的强度', '体育锻炼的强度'])
```

```python
# data_JobHomeSport.iloc[range(10)]
for column in data_JobHomeSport.columns:
    if column == 'ID':
        continue
    else:
        data_JobHomeSport[column] = data_JobHomeSport[column].map(replaceSport)
```

```python
for column in data_JobHomeSport.columns:
    if column == 'ID':
        continue
    else:
        temp = pd.DataFrame(data_JobHomeSport, columns=['ID', column])
        temp = temp.dropna()
        temp.to_excel(f'D:\\数学建模\\数模第四次作业\\A题-附件\\{column}.xlsx')
```

```python
data_doSports = pd.DataFrame(data, columns=['ID', '是否参加体育锻炼', '平均每天体育锻炼时间'])
data_doSports.insert(data_doSports.shape[1],'每周锻炼时间', 0)
def caculateSportTime(x):
    days = x['是否参加体育锻炼']
    avg_time = x['平均每天体育锻炼时间']
    if days == 1:
        return 0
    elif days == 2:
        return 1.5 * avg_time
    elif days == 3:
        return 4 * avg_time
    elif days == 4:
        return 5.5 * avg_time
    else:
        return 0
    
data_doSports['每周锻炼时间'] = data_doSports.apply(caculateSportTime, axis=1)
```

```python
data_doSports = data_doSports.drop(columns=[ '是否参加体育锻炼', '平均每天体育锻炼时间'])
data_doSports.to_excel('D:\\数学建模\\数模第四次作业\\A题-附件\\平均每周体育锻炼时间.xlsx')
```

```python
def replaceBreakfast(x):
    if x == 0:
        return "每天吃早餐"
    elif x in range(1, 3):
        return "偶尔不吃早餐"
    elif x in range(3, 6):
        return "经常不吃早餐"
    else:
        return "几乎或从不吃早餐"

data['不吃早餐'] = data['不吃早餐'].map(replaceBreakfast)
```
