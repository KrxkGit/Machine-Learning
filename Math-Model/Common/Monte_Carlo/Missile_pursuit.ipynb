```python
import torch
import math
from d2l import torch as d2l
import matplotlib.pyplot as plt
import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'
```

```python
device = d2l.try_gpu()
```

```python
# a为导弹，b为船
def plot_track(x_a0, y_a0, v_a ,a_range, x_b0, y_b0, v_bx, v_by, delta_t, endx, endy, auto_increase_range):
    xb = [float(x_b0)]
    yb = [float(y_b0)]
    xb = torch.arange(x_b0, endx, v_bx*delta_t, device=device)
    yb = torch.arange(y_b0, endy, v_by*delta_t, device=device)
    
    min_size = min(xb.shape, yb.shape)[0]
    xb = xb[:min_size]
    yb = yb[:min_size]
    ac_time = 0
    
    xa = [x_a0]
    ya = [y_a0]
    solve = None
    
    for i in range(min_size):
        ac_time += delta_t
        dxb = xb[i]
        dyb = yb[i]
        v_ax = (dxb - xa[-1])
        v_ay = (dyb - ya[-1])
        alpha = math.atan(v_ay/v_ax) # 求出速度角

        # 求出各方向移动距离
        v_ax = math.cos(alpha) * v_a * delta_t
        v_ay = math.sin(alpha) * v_a * delta_t
        
        a_range -= math.sqrt(v_ax**2 + v_ay**2)
        if a_range < 0 and not auto_increase_range:
            break
        xa.append(v_ax + xa[-1])
        ya.append(v_ay + ya[-1])
        
        if xa[-1] >= dxb and ya[-1] >= dyb:
            solve = (xa[-1], ya[-1])
            # 截断方便绘图
            xb = xb[:i]
            yb = yb[:i]
            break
        
        
    fig, ax = plt.subplots()
    
    # 定义颜色
    color_boat = 'blue'
    color_missile = 'orange'
    color_meet = 'red'
    
    plt.xlabel('x')
    plt.ylabel('y')
    
    ax.scatter(x_a0, y_a0, c=color_missile, label="Missile")
    ax.scatter(x_b0, y_b0, c=color_boat, label="Boat")
    if solve == None:
        print("未追上！")
    
       
    else:
        if a_range < 0:
             print("追上还需追加：", - int(a_range) + 1, " 单位射程。\n自动追加射程。")
        print("成功追上！\n追上时间:", ac_time,  "\n导弹坐标:", solve)
        ax.scatter(xa[-1], ya[-1], c=color_meet, label="Meet")
    
    ax.plot(xb.to('cpu').numpy(), yb.to('cpu').numpy(), label="Boat_trace", color=color_boat)
    ax.plot(xa, ya,  label="Missile_trace", color=color_missile)
    ax.legend()
    plt.title("Missile_pursuit")
    plt.show()
    return fig
```

```python
delta_t = 1e-3
vb = 1
va = 3 * math.sqrt(2)
fig = plot_track(-60, -4, va, 0, 80, 10, vb, vb, delta_t, 1e5, 1e5, auto_increase_range=True)
fig.savefig('Missle_pursuit.png')
```
